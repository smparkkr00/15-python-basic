```python
# while 조건문: 
#     반복할 코드

i = 0
while i < 5:
    print('hello world')
    i += 1
```

```python
# 초기값 1부터 시작하기

i = 1
while i <= 5:
    print('hello world', i)
    i += 1
```

```python
# 초기값 감소시키기

i = 5
while i > 0:
    print('hello world', i)
    i -= 1
```

```python
# 입력한 횟수대로 반복하기

count = int(input("반복할 횟수를 입력하세요:"))

i = 0
while i < count:
    print('hello world',i)
    i += 1
```
