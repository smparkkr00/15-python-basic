```python
# 리스트의 가장 작은 수, 큰 수, 합계 구하기

a = [38, 21, 53, 62, 19]

smallest = a[0]

for i in a:
    if i < smallest:
        smallest = i

print(smallest)
```

```python
# 리스트 컴프리헨션(comprehension) 사용하기

## 1. [식 for 변수 in 리스트]
## 2. list(식 for 변수 in 리스트)

a = [i for i in range(10)]
print("a:",a)

# for i in range(10):
#     print(i, end = "")

b = list(j for j in range(10))
print("b:",b)

c = [i + 5 for i in range(10)]
print("c:",c)

d = [i * 2 for i in range(10)]
print("d:",d)
```

```python
# 리스트 표현식에서 if 조건문 사용하기
## 1. [식 for 변수 in 리스트 if 조건식]
## 2. list(식 for 변수 in 리스트 if 조건식)

a = [i for i in range(10) if i % 2 == 0]
print(a)

b = [a + 5 for a in range(10) if a % 2 == 1]
print(b)
```

```python
# for문과 if문을 여러 번 사용하기

## [식 for 변수1 in 리스트1 
##    for 변수2 in 리스트2 ··· ]

a = [
    i * j
    for j in range(2, 10)
    for i in range(1, 10)
]

print(a)
```

```python
# 리스트에 map 사용하기

## list(map(함수, 리스트)) 
## tuple(map(함수, 튜플))

a = ["4", "2", "3", "1"]
for i in range(len(a)):
    a[i] = int(a[i])

print(a)
```

```python
a = [1.2, 2.5, 3.7, 4.6]

a = list(map(int, a))

print(a)
```
