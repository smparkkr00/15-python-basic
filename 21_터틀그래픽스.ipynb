```python
# 오각형 만들기
import turtle as t

t.shape("turtle")

t.clear()

for i in range(5):
    t.forward(300)
    t.right(144)
```

```python
# 육각형 만들기
import turtle as t

t.shape("turtle")

t.clear()

for i in range(6):
    t.forward(100)
    t.right(60)

```

```python
#다각형에 색칠하기

import turtle as t

n = 6 
t.shape('turtle')
t.color('red')
t.begin_fill()
for i in range(6):
    t.forward(100)
    t.right(60)
t.end_fill()
```

```python
import turtle as t

t.shape('turtle')
t.circle(120)
```

```python
# 원을 반복해서 그리기

import turtle as t

n=60
t.clear()
t.shape('turtle')
t.speed('fastest')
for i in range(n):
    t.circle(120)
    t.right(360/n)
```

```python
# 선으로 복잡한 무늬 그리기
import turtle as t

t.clear()
t.shape('turtle')
t.speed("fastest")

for i in range(300):
    t.forward(i)
    t.right(91)
```
