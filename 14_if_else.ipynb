```python
# else와 들여쓰기

x = 5

if x == 10:
    print("10입니다.")

else:
    print("x에 들어있는 숫자는")
    print("10이 아닙니다.")
```

```python
# if 조건문의 동작 방식 알아보기

if True:
    print("참")
else:
    print("거짓")

if False:
    print("참")
else:
    print("거짓")

if None:
    print("참")
else:
    print("거짓")
```

```python
# 10이상이고 20미만인 값은 참

x = int(input("숫자 입력:"))

if x >= 10 and x < 20:
    print("참")
    
else:
    print("거짓")

if 10 <= x < 20:
    print("참")
    
else:
    print("거짓")
```

```python
written_test = 75
coding_test = True

if written_test >= 80 and coding_test == True:
    print("합격")
else:
    print("불합격")
```

```python
kor, eng, math, sci = map(int, input().split(" "))


if all(0 <= x <= 100 for x in [kor, eng, math, sci]):
    if(kor + eng + math + sci)/4 >= 80:
        print("합격")

    else:
        print("불합격")
else:
    print("잘못된 점수")
```
