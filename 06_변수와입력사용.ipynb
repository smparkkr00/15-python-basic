```python
# 입력 값을 정수로 변환하기
a = int(input("첫 번째 숫자:"))
b = int(input("두 번째 숫자:"))
print(a + b)
```

```python
# 입력 값을 변수 두개에 저장하기
a, b = input("문자열 두개를 입력하세요:").split()

print(a)
print(b)
"가 나".split()
```

```python
# print("가 나".split())
a, b = ['가', '나']
print(a)
print(b)
a, b
```

```python
# 두 숫자의 합 구하기
a, b = input("숫자 2개 입력: ").split()
print(a)
print(b)
print(a + b)
```

```python
a, b = input("숫자 2개 입력: ").split()
a = int(a)
b = int(b)
print(a + b)
```

```python
a, b = map(int, input("숫자 2개 입력:").split())
c, d = map(int, ["10", "20"])
print(a + b)
print(c + d)
```

```python
# 입력받은 값을 콤마를 기준으로 분리하기
a, b = map(int, input("숫자 2개 입력:").split(","))
print(a + b)
```
