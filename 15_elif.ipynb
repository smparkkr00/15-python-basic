```python
# elif 사용하기

x = 20

if x == 10:
    print("10입니다")
elif x == 20:
    print("20입니다")
```

```python
# if, elif, else 모두 사용하기

x = 30

if x == 10:
    print("10입니다")
elif x == 20:
    print("20입니다")
else:
    print("아무것도 아닙니다")
```

```python
# 음료수 자판기 

button = int(input())

if button == 1:
    print("콜라")

elif button == 2:
    print("사이다")

elif button == 3:
    print("환타")

else:
    print("상품 없음")
```

```python
x = int(input())



if 11 <= x <= 20:
    print("11~20")

elif 21 <= x <= 30:
    print("21~30")

else:
    print("아무것도 해당되지 않음")
```

```python
age = int(input())
balance = 9000

if age < 7:
    print("다시 입력하세요")
    
if 7 <= age <= 12:
    print(f"{balance - 650}")
elif 13 <= age <= 18:
    print(f"{balance - 1050}")
elif age >= 19:
    print(f"{balance - 1250}")
```
