```python
# 값을 여러개 출력하기

print(1, 2, 3)
```

```python
# sep로 값 사이에 문자넣기

print(1, 2, 3, sep = " + ")
```

```python
print(1920, 1000, sep = "x")
```

```python
# 줄바꿈 활용
print(1, 2, 3, sep = "\n")
```

```python
print("1\n2\n3")
```

```python
# end 사용하기
print(1, end = " ")
print(2, end = " ")
print(3)
```

```python
print(1, end = " ")
print(2, end = " ")
print(3)

print(1, 2, 3)
```
