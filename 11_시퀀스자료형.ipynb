```python
# 인덱스 사용하기

a = [31, 50, 79, 80, 15]

print(a[0], a[3])

b = (30, 21, 53, 62, 19)

print(b[0], b[4])

c = range(0,10,2)
print(c[0], c[2])

d = "Hello, World!"
print(d[7], d[0])
```

```python
# 음수 인덱스 지정하기

a = [38, 21, 53, 62, 19]

print(a[-1], a[-5])

b = (30, 21, 53, 62, 19)
print(b[-2])

c = range(0,10,2)
print(c[-1])

d = "Hello, World!"
print(d[-1], d[-4])
```

```python
# 리스트 요소에 값 할당하기

a = [0, 0, 0, 0, 0]
a[0] = 38
a[1] = 21
a[2] = 53
a[3] = 62
a[4] = 19

print(a)
```
