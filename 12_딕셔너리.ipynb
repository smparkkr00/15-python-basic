```python
# 빈 딕셔너리 만들기

x = {}
y = dict()

print(x)
print(type(x))
print(bool(x))
print(y)
print(type(y))
print(bool(y))
```

```python
# dict 함수로 딕셔너리 만들기

lux1 = dict(health=490, mana=334, melee=550, armor=18.72)
print(lux1)
print(type(lux1))

lux2 = dict(zip(['health', 'mana', 'melee', 'armor'],[490, 334, 550, 18.72]))
print(lux2)
print(type(lux2))

lux3 = dict([('health', 490), ('mana', 334), ('melee', 550), ('armor', 18.72)])
print(lux3)
print(type(lux3))
```

```python
# 딕셔너리의 키에 접근하고 값 할당하기

lux = {
    "health": 490,
    "mana": 334,
    "melee": 550,
    "armor": 18.72
}

print(lux["health"])
print(lux["armor"])
lux["health"] = 2037
lux["mana"] = 1184
print(lux)
```

```python
lux["mana_regen"] = 3.28
print(lux)
```

```python
# 딕셔너리에 키가 있는지 확인하기

print('health' in lux)
print('attack_speed' in lux)
```

```python
# 딕셔너리의 키 개수 구하기

print(lux)
print(len(lux))
```
